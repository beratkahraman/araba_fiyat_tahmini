```python
import pandas as pd
import os

data_path = os.path.join("..", "data", "raw", "ilanlar_ham.csv")
df = pd.read_csv(data_path, low_memory=False)
df.head()
```

```python
df.info()
```

```python
import numpy as np

df['Fiyat'] = df['Fiyat'].astype(str).str.replace(' TL', '').str.replace('.', '')
df['Fiyat'] = pd.to_numeric(df['Fiyat'], errors='coerce')

print(f"Temizlik öncesi satır sayısı: {len(df)}")
df.dropna(subset=['Fiyat'], inplace=True)
print(f"Fiyatı olmayanlar silindikten sonra satır sayısı: {len(df)}")

df['Fiyat'] = df['Fiyat'].astype(int)

df.head()
```

```python
df['Kilometre'] = df['Kilometre'].astype(str).str.replace(' km', '').str.replace('.', '')

df['Kilometre'] = pd.to_numeric(df['Kilometre'], errors='coerce')

df.dropna(subset=['Kilometre'], inplace=True)

df['Kilometre'] = df['Kilometre'].astype(int)

df[['Marka', 'Model', 'Fiyat', 'Kilometre']].head()
```

```python
df['Yıl'] = pd.to_numeric(df['Yıl'], errors='coerce')
df.dropna(subset=['Yıl'], inplace=True)
df['Yıl'] = df['Yıl'].astype(int)
```

```python
gereksiz_sutunlar = ['Başlık', 'İlan No', 'İlan Tarihi']

mevcut_gereksizler = [col for col in gereksiz_sutunlar if col in df.columns]
df.drop(columns=mevcut_gereksizler, inplace=True)

df.head()
```

```python
df.describe().T
```

```python
# 1. Kilometre Filtresi: 1.000.000 km üzeri araçları hatalı kabul edip silelim
df = df[df['Kilometre'] < 1000000]

# 2. Fiyat Filtresi: 
# Sınır: 50.000 TL (Bunun altı genelde hasarlı/hurda veya hatalı girilmiştir)
df = df[(df['Fiyat'] > 50000)]

print(f"Aykırı değerler temizlendikten sonra kalan satır sayısı: {len(df)}")
```

```python
df['Tramer'] = df['Tramer'].astype(str)

df['Tramer'] = df['Tramer'].replace(['nan', 'None', 'Belirtilmemiş'], '0')

df['Tramer'] = df['Tramer'].str.replace(' TL', '').str.replace('.', '')

df['Tramer'] = pd.to_numeric(df['Tramer'], errors='coerce').fillna(0).astype(int)

print(df['Tramer'].describe())
```

```python
output_path = os.path.join("..", "data", "processed", "ilanlar_final.csv")

os.makedirs(os.path.dirname(output_path), exist_ok=True)

df.to_csv(output_path, index=False, encoding='utf-8-sig')

print(f"Temiz veri seti başarıyla kaydedildi: {output_path}")
```
